````python
from langchain_huggingface import HuggingFaceEmbeddings
huggingface_embeddings=HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
````

````python
result=huggingface_embeddings.embed_query("India is a growing country")

print("Result: ", result)
print("-"*50)
print("Length: ", len(result))
````

````python
documents=["what is a capital of USA",
           "who is a president of usa",
           "who is a prime minister of india"]

result=huggingface_embeddings.embed_documents(documents)

# Print the total length of the result
print("Total Number of Embedded Sentences: ", len(result))
print("-" * 50)

# Print the embedding vectors for all sentences
print("Embedding Vectors for All Sentences:")
print(result)
print("=" * 150)

# Print the length of the embedding vector for one sentence
print("Length of Embedding Vector for a Single Sentence: ", len(result[0]))
print("-" * 50)

# Print the embedding vector for one sentence
print("Embedding Vector for the First Sentence:")
print(result[0])
````

````python
documents=["what is a capital of USA",
           "who is a president of usa",
           "who is a prime minister of india"]

my_query="narendra modi is indian prime minister"

aquery_embedding=huggingface_embeddings.embed_query(my_query)
adocument_embedding=huggingface_embeddings.embed_documents(documents)
````

```python
# The cosine similarity scores indicate how similar the query is to each document.
# The resulting array is as follows:
# array([[-0.0682291 ,  0.23930707,  0.7199079 ]])
# 
# The third score (0.7199079) is the highest, suggesting that the third document
# ("who is a prime minister of india") has the greatest similarity to the query
# ("narendra modi is indian prime minister") compared to the other documents.
```


````python
from sklearn.metrics.pairwise import cosine_similarity

scores=cosine_similarity([aquery_embedding],adocument_embedding)
scores
````
